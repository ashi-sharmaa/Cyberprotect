```python
import math
import statistics
import numpy as np
import scipy.stats
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (8.0, 6.0)
```

```python
import io
dataset = pd.read_csv(io.BytesIO(uploaded['AnonymityPoll.csv']))
#remove unnecessary columns
dataset.drop(['Smartphone', 'Sex', 'State', 'Region', 'Conservativeness'], axis = 1, inplace = True)
#filter out those who don't use the internet
users = dataset[dataset['Internet.Use'] > 0]
#this shows us the public's average rating of privacy importance and amount of personal info online
print(users.describe())
#selects the users who have tried to stay anonymous but believe it isn't possible
anonymity = users[(users['Anonymity.Possible'] == 1) & (users['Tried.Masking.Identity'] == 0)]
print(anonymity.shape)
#sample size of about 1/3 of internet users polled, put into a line graph
users2 = users.head(250)
plt.figure()
plt.plot(users2['Age'], users2['Privacy.Importance'])
plt.xlabel('Age')
plt.ylabel('Privacy Importance (0-100)')
plt.title('Age vs. Privacy Importance Opinion', fontsize = 22)
plt.show()
```
